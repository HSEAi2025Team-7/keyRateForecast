```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import MaxAbsScaler
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

CountVectorizer + LogisticRegression

```python
df = pd.read_csv("df_key_rate.csv")

x_train, x_test, y_train, y_test = train_test_split(df.text, df.target, random_state=42)

vec = CountVectorizer(ngram_range=(1,1))
bow = vec.fit_transform(x_train)
bow_test = vec.transform(x_test)

scaler = MaxAbsScaler()
bow = scaler.fit_transform(bow)
bow_test = scaler.transform(bow_test)


clf = LogisticRegression(max_iter=200, random_state=42)
clf.fit(bow, y_train)
pred = clf.predict(bow_test)
print(classification_report(y_test, pred))

f1_macro_cv_1g = f1_score(y_test, pred, average='macro')
print(f"F1 macro: {f1_macro_cv_1g:.2f}")
```

```python
results_df = pd.DataFrame(columns=["model", "f1_macro"])

results_df.loc[len(results_df)] = {
    "model": "CountVectorizer(1gramm) + LogisticRegression",
    "f1_macro": f1_macro_cv_1g
}
results_df
```

```python
vec = CountVectorizer(ngram_range=(1, 3))
bow = vec.fit_transform(x_train)
bow_test = vec.transform(x_test)


scaler = MaxAbsScaler()
bow = scaler.fit_transform(bow)
bow_test = scaler.transform(bow_test)

clf = LogisticRegression(max_iter=200, random_state=42)
clf.fit(bow, y_train)
pred_thrgramm = clf.predict(bow_test)
print(classification_report(y_test, pred_thrgramm))

f1_macro_cv_3g = f1_score(y_test, pred_thrgramm, average='macro')
print(f"F1 macro: {f1_macro_cv_3g:.2f}")
```

```python
results_df.loc[len(results_df)] = {
    "model": "CountVectorizer(1-3gramm) + LogisticRegression",
    "f1_macro": f1_macro_cv_3g
}
results_df
```

TfidfVectorizer + LogisticRegression

```python
vec = TfidfVectorizer(ngram_range=(1, 1))
vec_train = vec.fit_transform(x_train)
vec_test = vec.transform(x_test)


scaler = MaxAbsScaler()
vec_train = scaler.fit_transform(vec_train)
vec_test = scaler.transform(vec_test)

clf = LogisticRegression(max_iter=200, random_state=42)
clf.fit(vec_train, y_train)
pred_tfidf_1g = clf.predict(vec_test)
print(classification_report(y_test, pred_tfidf_1g))

f1_macro_tfidf_1g = f1_score(y_test, pred_tfidf_1g, average='macro')
print(f"F1 macro: {f1_macro_tfidf_1g:.2f}")
```

```python
results_df.loc[len(results_df)] = {
    "model": "TfidfVectorizer(1gramm) + LogisticRegression",
    "f1_macro": f1_macro_tfidf_1g
}
results_df
```

```python
vec = TfidfVectorizer(ngram_range=(1, 3))
vec_train = vec.fit_transform(x_train)
vec_test = vec.transform(x_test)


scaler = MaxAbsScaler()
vec_train = scaler.fit_transform(vec_train)
vec_test = scaler.transform(vec_test)

clf = LogisticRegression(max_iter=200, random_state=42)
clf.fit(vec_train, y_train)
pred_tfidf_3g = clf.predict(vec_test)
print(classification_report(y_test, pred_tfidf_3g))

f1_macro_tfidf_3g = f1_score(y_test, pred_tfidf_3g, average='macro')
print(f"F1 macro: {f1_macro_tfidf_3g:.2f}")
```

```python
results_df.loc[len(results_df)] = {
    "model": "TfidfVectorizer(1-3gramm) + LogisticRegression",
    "f1_macro": f1_macro_tfidf_3g
}
results_df
```

```python
results_df.to_csv("results_df.csv", index=False)
```
